```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("Breast Cancer Data Set.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns = "Unnamed: 32", axis=1)
```

```python
x = df.iloc[:,3:-1].values
y = df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn .preprocessing import LabelEncoder

le = LabelEncoder()
le2 = LabelEncoder()

x[:,1] = le.fit_transform(x[:,1])
x[:,2] = le.fit_transform(x[:,2])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
df.info()
```

```python
from sklearn .preprocessing import StandardScaler

sc = StandardScaler()

xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
# 1.Initilize the ANN Model

model = Sequential()

# 2.Pass hidden layer

model.add(Dense(units = 6, activation = "relu"))

# 3.Pass Activation Funcation

model.add(Dense(units = 1, activation = "sigmoid"))

# 4.To Establish the connection between layers

model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])

# 5.Train the model

model.fit(xtrain, ytrain, batch_size = 30, epochs = 100)

# 6.Take Prediction

ypred = model.predict(xtest)

# 7.Set Threshhold Value

ypred = ypred>0.5

# 8.Get classification report

print(classification_report(ytest,ypred))
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(xtrain, ytrain)

ypred = logreg.predict(xtest)

print(classification_report(ytest, ypred))
```
